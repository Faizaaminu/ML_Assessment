**Linear Regression** is building a model that can estimate a linear relationship between one or more independent variables.  


To train this model, we need to split the dataset into:  
- **Training set** = 80  
- **Testing set** = 20  

The **Training Set** is the data that we provide when training the model, while the **Test Set** is used to evaluate the accuracy of the model after training.  

The reason we have to split the dataset into training and test sets is that if we test the model on the same data it was trained on, we wouldn’t get a realistic picture of how well the model performs on unseen data. The whole idea is to input years of experience and get the predicted salaries.  

The most common way to split the data is **80/20**—80% for training and 20% for testing. This ratio can vary depending on the size of the dataset.  

When the model is well-trained, We should have:  
- **m = 9332.9** (Slope, θ1)

We want the output to be in a **matrix format** as a **two-dimensional NumPy array**.

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

## lets try to visualize our data on the train set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
print(X_train.shape) #this is a numpy array that features of 80 rows and 1 columns.
print(X_test.shape) #this is a matrix with two dimentional
print(y_train.shape) #this is a vector
print(y_test.shape) #this is a vector
```

```python
from sklearn.linear_model import LinearRegression
#create the model
model = LinearRegression()

#train the model using the traning set
model.fit(X_train, y_train) #behind the scrence this is training your model i.e finding the slop for the feature/object
```

```python
#get the model predict on the test set
y_pred = model.predict(X_test)
print(y_pred)
print(y_test)
```

```python
#visualizing the training set with the predictions
plt.scatter(X_train, y_train, color = 'pink')
plt.plot(X_train, model.predict(X_train), color ="red")
plt.title("Salary vs Years of Experience (Training set)")  #title
plt.xlabel("Years of Experience")  # Label for x-axis
plt.ylabel("Salary (USD)")  # Label for y-axis
plt.show()
```

```python
#visualizing the test set with the predictions
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_test, model.predict(X_test), color ="blue")
plt.title("Salary vs Years of Experience (Test set)")  #title
plt.xlabel("Years of Experience")  # Label for x-axis
plt.ylabel("Salary (USD)")  # Label for y-axis
plt.show()
```

## let's consider other metrics as well.

```python
from sklearn.metrics import mean_squared_error, r2_score
```

In linear regression, these two metrics help evaluate how well the model predicts the data:  

1. **Mean Squared Error (MSE)**:  
   - Measures how far the predicted values are from the actual values.  
   - It calculates the average of the squared differences between actual and predicted values.  
   - A lower MSE means the model's predictions are more accurate.  

2. **R² Score (R-squared)**:  
   - Shows how well the model explains the variation in the data.  
   - It ranges from 0 to 1, where 1 means perfect prediction and 0 means the model doesn't explain the data at all.  
   - Higher R² means a better fit.  

NB:  
- **MSE tells us how wrong our predictions are.**  
- **R² tells us how well our model fits the data.**  

```python
#What is error? This is the difference between the actual values and the predicted values
#i.e the difference between the y_test and y_pred.
print("Mean Squared error: ", mean_squared_error(y_test, y_pred))
print("R2 Score: ", r2_score(y_test, y_pred)) #this need to be close to one as possible.
#remember we have a small dataset with 80 point
```

```python
print('Slop: ', model.coef_)
print('Intercept: ', model.intercept_)

#this is our model. we can take the slop and intercept to create a simple line using X is equal MX plus b.
#with that we can get the best fit line for the data
```

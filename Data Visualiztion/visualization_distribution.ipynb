## Data Visualization

```python
plt.figure(figsize=(10, 6))
# Plotting the distribution of species
sns.countplot(x='species', data=iris_df, palette='Set2',hue='species')
plt.title('Distribution of Iris Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

```python
# Plotting the distribution of each feature
plt.figure(figsize=(12, 8))
for i, column in enumerate(iris_df.columns[:-1], 1):
    plt.subplot(2, 2, i)
    sns.histplot(iris_df[column], kde=True, bins=30)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

1. Sepal Length: most values lie between 5.0 to 6.5 cm, with peak around 5.5cm
2. Sepal Widtrh: Distribution is slightly skewed left; many values cluster between 2.8 to .3.5cm
3. epetal Length: strong peak between 1.0 and 5.0 cm; a wide range due clear species seperation 
4. Petal Width: Bimodal distribution, suggesting species specific charatreristic

 ### Hist Plot : Iris Species Sepal  & Petal

```python
plt.figure(figsize=(12,6))
# Create FacetGrid with Seaborn
sns.FacetGrid(iris_df,hue='species',palette='viridis', height=5) \
    .map(sns.histplot, 'sepal_length') \
    .add_legend();
plt.title("Distribution of Sepal Length")
plt.show();
```

```python
plt.figure(figsize=(12,6))
# Create FacetGrid with Seaborn
sns.FacetGrid(iris_df,hue='species',palette='viridis', height=5) \
    .map(sns.histplot, 'sepal_width') \
    .add_legend();
plt.title("Distribution of Sepal    Width")
plt.show();
```

Sepal Length:
- The distribution appears approximately normal.
- The majority of sepal lengths fall between 4.3 cm and 7.9 cm.
- There are some noticeable outliers in virginica .
- The overlap between Versicolor and Virginica based on sepal length may be more challenging compared to Setosa.
- This indicates that there is less clear separation between Versicolor and Virginica species in terms of these features.<br>

```python
#Distplot
# Create FacetGrid with Seaborn
sns.FacetGrid(iris_df,hue='species', height=5) \
    .map(sns.histplot, 'petal_length') \
    .add_legend();
plt.title("Distribution of Petal Length")
plt.show();
```

Petal Length:
- Setosa flower can be linearly separable
- while versicolor and virginica distributions are overlapping
- The distribution shows a bimodal pattern, with peaks around 1.0 cm and 6.9 cm.
- Most petal lengths range from 1.0 cm to 6.9 cm.

```python
#Distplot
# Create FacetGrid with Seaborn
sns.FacetGrid(iris_df,hue='species', height=5) \
    .map(sns.histplot, 'petal_width') \
    .add_legend();
plt.title("Distribution of Petal Width")
plt.show();
```

1. Setosa species
   - has distinctly smaller petal length and width
   - Narrow range with high frequency at the lower end of values.
  2. Versicolor and Virginica:

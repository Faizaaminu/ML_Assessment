```python
#Filter for snow goose
snow_geese=birds[birds['Name'].str.contains("Snow goose", case=False)]
snow_geese
```

```python
snow_geese.describe()
```

<br>Bar Plot:Max Wingspan vs Name</br>

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=snow_geese, x='MaxWingspan', y='Name',hue='MaxWingspan',palette='viridis')
plt.title('Max Wingspan of Snow Goose')
plt.xlabel('Max Wingspan (cm)')
plt.ylabel('Bird Name')
plt.xticks(rotation=45)
plt.show()
```

Line Plot: MinLength vs Maxlength

```python

plt.figure(figsize=(10,6))
plt.plot(['MinLength','MaxLength'],
         [snow_geese['MinLength'].values[0] ,snow_geese['MaxLength'].values[0]],  marker='o', linestyle='-',
          color='b')
plt.title('Body Length Range of Snow Goose')
plt.ylabel('Length(cm)')
#plt.grid(True)
sns.set_style('darkgrid')
plt.show()
```

MinBody vs MaxBody

```python

plt.figure(figsize=(10,6))
plt.plot(['Min Body  Mass','Max Body Mass'],
         [snow_geese['MinBodyMass'].values[0] ,snow_geese['MaxBodyMass'].values[0]],  marker='s', linestyle='--',
          color='g')
plt.title('Body Mass Range of Snow Goose')
plt.ylabel('Mass(g)')
plt.grid(True)
plt.show()
```

Scatter Plot: Weight vs Wingspan

```python
# Scatter plot:Min Body Mass vs Max Wingspan
plt.figure(figsize=(8,5))
plt.scatter(snow_geese['MinBodyMass'],['MaxWingspan'], color='darkgreen', s=100)
plt.title('Snow Goose: Min Body Mass vs Max Wingspan')
plt.xlabel('MinBodyMass (g)')
plt.ylabel('MaxWingspan')
plt.grid(True)
plt.tight_layout()
plt.show()
```

length vs wingspan:compare with other geese

```python
# Scatter plot:Min Body Mass vs Max Wingspan
geese=birds[birds['Genus']== 'Anser']
plt.figure(figsize=(8,5))
sns.scatterplot(data=geese,x='MaxLength',y='MaxWingspan', hue='Name', palette='Blues_d', s=100)
plt.title('Snow Goose: Max length vs Max Wingspan')
plt.xlabel('MaxLength (g)')
plt.ylabel('MaxWingspan')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

#### Interpretation
- The maximum wingspan of the snow goose its large capability
- there's a clear range in body length, which may vary by subspecies.
- the scatter plot reveal the relation between body mass and wingspan, which helps understand flight adaptaions
